```python
# 初回のみライブラリをインポートする
import pandas as pd

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")

```

```python
def tax(x):
    return x * 1.10

df['単価'].apply(tax)
```
